```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importando base de dados com diversas informações soltadas pelo Metacritic (ano desconhecido)
df = pd.read_csv("games_sales.csv")
```

```python
display(df.info())
display(df)
```

```python
#Substituindo a Plataforma "2600" por "Atari 2600"
df["Platform"] = ["Atari 2600" if plat == "2600" else plat for plat in df["Platform"]]
df.fillna("--",inplace=True)
display(df)
```

```python
df.sort_values(by="Name", ascending=True, inplace=True)
linha = df.index[df["Name"] == "--"]
df.drop(linha, axis=0, inplace=True)
display(df.shape)
```

```python
#Apagando colunas inúteis para a análise e salvando um novo DF
games_df = df.drop(["User_Score", "User_Count", "Critic_Count", "Rating", "Other_Sales", "JP_Sales"], axis=1)
display(games_df.info())
```

```python
#Jogo desenvolvidos pela Blizzard
display(games_df.loc[games_df["Developer"] == "Blizzard Entertainment"])
```

```python
#Mostrando apenas jogos publicados pela Nintendo e pela Sega
display(games_df.loc[games_df["Publisher"].isin(["Nintendo", "Sega"])])
```

```python
#Exibindo jogos da Nintendo que foram lançados para GBA e GB
display(games_df.loc[(games_df["Publisher"] == "Nintendo") & (games_df["Platform"].isin(["GBA", "GB"]))])
```

```python
#Agrupando e contando os jogos por plataforma
platforms = games_df.groupby(by=["Platform"]).count()
platforms.reset_index(inplace=True)
platforms.drop(["Genre", "NA_Sales", "EU_Sales", "Global_Sales", "Critic_Score", "Developer", "Year_of_Release", "Publisher"], axis=1, inplace=True)
platforms.rename(columns={"Name" : "Games/Platform"}, inplace=True)
platforms.set_index("Platform", inplace=True)
display(platforms)
display(platforms.info())
```

```python
#Gerando gráfico das plataformas e a quantidade de jogos
plt.figure(figsize=(20,10))

platforms.plot(kind="barh", 
    ylabel="Platforms", 
    xlabel="Games Available", 
    color="purple", 
    title="")

sns.despine()
plt.grid(axis="x")
plt.show()
```
